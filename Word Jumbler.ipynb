```python
from tkinter import * 
from tkinter.ttk import * 
from time import strftime 
from tkinter import messagebox
import random

root = Tk()
root.title("Identify the phrase")
root.geometry("1200x710")


#getting the files containing 
#the phrases

with open('phrases.txt') as f:
    phrases = f.read().splitlines() 
f.close() 
random.shuffle(phrases)

def generate_jumbled_phrases(var):
    #removing the spaces and the end punctuation
    #and storing them to add them later
    list3=[]
    for i in range(len(var)):
        if var[i]==' ':
            list3.append(i)
        
        
    var1=var.replace(' ','')
    var1=var1.replace(var1[len(var1)-1],'')
    list1=[]
    list1[:0]=var1 
    list2=[]

    #scibling them
    while (len(list1)!=0):
        c=random.choice(list1)
        list2.append(c)
        list1.remove(c)

    list2.append(var[len(var)-1])   
    for x in list3:
        list2.insert(x,' ')
    
    str1 = ''.join(list2)
    return str1

#setting the default time of the game
timeleft=30

#options for changing time
def change_time():
    #window for changing game time
    top = Toplevel()
    #changing time options
    r=IntVar()
    
    def update_time(value):
        global timeleft
        timeleft+=int(value)*10
        timeLabel.config(text= "Time left: " + str(timeleft))
        okay_button=Button(top,text='Done!',command=top.destroy)
        okay_button.pack()
        
    
    Radiobutton(top,text='30 seconds',variable=r, value=0,command= lambda: update_time(r.get())).pack()
    Radiobutton(top,text='40 seconds',variable=r, value=1,command= lambda: update_time(r.get())).pack()
    Radiobutton(top,text='50 seconds',variable=r, value=2,command= lambda: update_time(r.get())).pack()
    Radiobutton(top,text='1 minute',variable=r, value=3,command= lambda: update_time(r.get())).pack()
    
    

#initializing the score
score=0

#intialing the phrase
p=0

#Starting the game
def start_game(event):
    
    #start the counting
    if timeleft==30 or timeleft==40 or timeleft==50 or timeleft==60:
        countdown()
        
    next_phrase()
    
#Function for changing 
#the phrases after a fixed time
def next_phrase():
    global timeleft 
    global p
    global phrases
  
    # if a game is currently in play 
    if timeleft > 0: 
  
        # make the text entry box active. 
        e.focus_set() 
        e.delete(0, END) 
        
        # change the phrase  
        label.config(text = generate_jumbled_phrases(phrases[p]))
 
  
  
 #Countdown timer function  
def countdown():
                     
  
    global timeleft 
  
    # if a game is in play 
    if timeleft > 0: 
  
        # decrement the timer. 
        timeleft -= 1
          
        # update the time left label 
        timeLabel.config(text = "Time left: "
                               + str(timeleft)) 
                                 
        # run the function again after 1 second. 
        timeLabel.after(1000, countdown) 
        
    elif timeleft==0:
        messagebox.showinfo("Game Ended","Your score is" + str(score))
    
def match_text():
    global score
    global p
    global phrases
     
                     
     #check if it is correct
     #and if it is then increment
     #the score
    if e.get()==phrases[p]:
         score+=1
    
    scoreLabel.config(text = "Score: " + str(score))
    p+=1
    next_phrase()                
                     
                     
#building GUI for it

menubar = Menu(root) 
  
# Adding File Menu and commands 
file = Menu(menubar, tearoff = 0) 
menubar.add_cascade(label ='Options', menu = file) 
file.add_command(label ='Change Game Time', command = change_time) 
file.add_command(label ='Change Font', command = None)  
file.add_separator() 
file.add_command(label ='Exit', command = root.destroy) 
  
# display Menu 
root.config(menu = menubar) 



 # add an instructions label 
instructions = Label(root, text = "Type in the phrase you think it is",font = ('Helvetica', 24)) 
instructions.pack()  
  
# add a score label 
scoreLabel = Label(root, text = "Press Enter to start",font = ('Helvetica', 12)) 
scoreLabel.pack() 
  
# add a time left label 
timeLabel = Label(root, text = "Time left: " + str(timeleft), font = ('Helvetica', 12)) 
                
timeLabel.pack()                   

label=Label(root,font = ('Helvetica', 60))
label.pack()


#add the space for text to be entered                     
e=Entry(root)
e.pack()

#add a button to check
check_button=Button(root,text="Check!",command=match_text)
check_button.pack()


#run the 'start_game' function
#when the enter key is pressed                     
root.bind('<Return>', start_game)                     
e.pack()
                     
e.focus_set()  

                     
root.mainloop()



```
